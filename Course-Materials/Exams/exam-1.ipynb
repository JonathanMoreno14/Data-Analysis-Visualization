__Question 1:__ Describe the distribution of six-year graduation rates among large universities in Texas.  Which universities have the highest graduation rates?  The lowest?  What does the "shape" of the distribution look like?  Are there any outliers?  

```python
import pandas as pd
import seaborn as sns

txschools = pd.read_csv("txcolleges2016.csv")
sns.set_style("darkgrid")

txschools_sorted = txschools.sort_values(by = 'grad_rate', ascending = False)
txschools_sorted.columns
cols_to_keep = ['instnm', 'grad_rate']
txschools_sorted = txschools_sorted[cols_to_keep] 
txschools_sorted.columns = ['Texas Universities', 'Graduation Rates']
```

### Descriptive distribution:

```python
txschools_sorted = txschools.sort_values(by = 'grad_rate', ascending = False)
sns.stripplot(x = 'grad_rate', y = 'instnm', data = txschools_sorted)
```

The distribution of graduation rates among Texas universities is distributed in a positive correlation. The graduation rates range from 0.9 (Max) to  0.162 (Min) with a range of 0.7. The outlier is Rice University with a gradtuation rate at 0.9. The distribution seesm to be skewed to the right.
